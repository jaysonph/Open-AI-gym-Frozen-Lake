```python
import gym
import numpy as np
```

```python
def play_episodes(env, n_episodes, policy, random=False):
    wins = 0
    total_reward = 0
    
    for episode in range(n_episodes):
        state = env.reset()
        done = False
        while not done:
            
            if random:
                action = env.action_space.sample()
            else:
                action = policy[state]
            
            next_state, reward, done, info = env.step(int(action))
            
            env.render()
            
            total_reward += reward
            state = next_state
            
            if done and reward == 1.0:
                wins += 1
    
    average_reward = total_reward/n_episodes
    
    return wins, total_reward, average_reward
```

## Value Iteration

```python
def one_step_lookahead(env, V, state, discount_factor):
    action_values = np.zeros(env.nA)
    for action in range(env.nA):
        for prob, next_state, reward_sa, info in env.P[state][action]:
            action_values[action] += prob * (reward_sa + (discount_factor*V[next_state]))
    return action_values
    
def update_policy(env, V, policy, discount_factor):
    for state in range(env.nS):
        action_values = one_step_lookahead(env, V, state, discount_factor)
        policy[state] = np.argmax(action_values)
    return policy

def value_iteration(env, discount_factor=0.99999, max_iter=1000):
    V = np.zeros(env.nS)
    ref_V = V.copy()
    for num_iter in range(max_iter):
        prev_V = V.copy()
        for state in range(env.nS):
            action_values = one_step_lookahead(env, prev_V, state, discount_factor)
            V[state] = np.max(action_values)   # Best Value at one state
        if num_iter%20==0:
            if (np.all(np.isclose(V, ref_V))):
                break
            ref_V = V.copy()
    policy = np.zeros(env.nS)
    policy = update_policy(env, V, policy, discount_factor)
    print('=============== V.I. Ends at Iteration {} ==============='.format(num_iter))
    print('Optimal V = ', V)
    print('Optimal policy = ', policy)
    return V, policy
```

```python
env = gym.make('FrozenLake-v0')
env.reset()
opt_V, opt_policy = value_iteration(env)
```

```python
wins, total_reward, avg_reward = play_episodes(env, 100, opt_policy)
```

```python
print('===================== Value iteration =====================')
print('Total Wins: ', wins)
print('Total Reward: ', total_reward)
print('Average Reward: ', avg_reward)
```

## Policy iteration

```python
def policy_eval(env, V, policy, discount_factor):
    policy_value = np.zeros(env.nS)
    for state, action in enumerate(policy):
        for prob, next_state, reward_next_state, info in env.P[state][action]:
            policy_value[state] += prob * (reward_next_state + (discount_factor*V[next_state]))
    return policy_value

def policy_iteration(env, discount_factor=0.999, max_iter=1000):
    V = np.zeros(env.nS)
    policy = np.random.randint(0,4, env.nS)
    prev_policy = policy.copy()
    for num_iter in range(max_iter):
        V = policy_eval(env, V, policy, discount_factor)
        policy = update_policy(env, V, policy, discount_factor)
        if num_iter % 30 == 0:  #  longer inspection period --> learn better (avoiding short-term fluctuation)
            if (np.all(np.equal(policy, prev_policy))):
                break
            policy_prev = policy.copy()
    print('=============== P.I. Ends at Iteration {} ==============='.format(num_iter))
    print('Optimal V = ', V)
    print('Optimal policy = ', policy)
    return V, policy
```

```python
env = gym.make('FrozenLake-v0')
env.reset()
opt_V, opt_policy = policy_iteration(env)
```

```python
n_episode = 10
env.reset()
wins, total_reward, avg_reward = play_episodes(env, n_episode, opt_policy)
```
